```python
#Print the sum of all even & odd numbers in a range separately
n = int(input("Enter number : "))
even_sum = 0
odd_sum = 0

for i in range(1, n + 1):
  if i % 2 == 0:
    even_sum += i
  else:
    odd_sum += i

print(f"Sum of even numbers: {even_sum}")
print(f"Sum of odd numbers: {odd_sum}")
```

```python
# Print all the factors of a number
n = int(input("Enter a number to find its factors: "))

print(f"The factors of {n} are:")
for i in range(1, n + 1):
  if n % i == 0:
    print(i)
```

```python
# Accept a number and check if it a perfect number or not.
n = int(input("Enter number : "))
sum = 0

for i in range (1,n):
  if n % i == 0 :
    sum = sum + i

if sum == n :
  print("Perfect number")
else :
  print("Not a perfect number")
print(sum)
```

```python
#Check wether the number is prime or not
n = int(input("Enter number : "))
count = 0

for i in range(1,n+1):
  if n % i == 0 :
    count = count + 1

if count == 2 :
  print("Prime number")
```

```python
#Reverse a string without using in build functions.
n = input("Enter a string : ")
b = ""
for i in range (len(n)-1,-1,-1):
  b = b + n[i]

print(b)
```

```python
#Check string is Pallindrome or not
n = input("Enter string : ")
b = ""

for i in range (len(n)-1,-1,-1):
  b = b + n[i]

if n == b :
  print("Pallindrome")
else :
  print("Not a Pallindrome")
```

```python
#Count all letters, digits, and special symbols from a given string
'''Given: str1 = "P@#yn26at^&i5ve"
Expected Outcome:
Total counts of chars, digits, and symbols
Chars = 8
Digits = 3
Symbol = 4'''

n = 'P@#yn26at^&i5ve'
char = 0
dig = 0
schar = 0

for i in n :
  if i.isdigit():
    dig = dig + 1
  elif i.isalpha():
    char = char + 1
  else :
    schar = schar + 1

print(dig)
print(char)
print(schar)
```
