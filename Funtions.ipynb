```python
# the thing you accept is parameter and the thing you provide is arguments
def sum(a,b):
  print(f"the sum of your number is {a+b}")

sum(3,2)
sum(45,2)
```

```python
# positional argument
def sum(a,b):
  print(f"the sum of your number is {a+b}")

sum(3,8)

#keyword arguments
def intro(name,age):
  print(name)
  print(age)

intro(name = "Snehal",age = 19)

#default argument
def add(a,b=10):
  print(a+b)

sum(12,2)
```

```python
#Check string is palindrome or not
def palindrome(str):
  rev = ""
  for i in range(len(str)-1,-1,-1):
    rev = rev + str[i]

  if rev.lower() == str.lower():
    print("palindrome")
  else:
    print("not palindrome")

palindrome("Naman")
```
